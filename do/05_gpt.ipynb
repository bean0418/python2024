```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Data: call handling times
data = np.array([6, 26, 8, 2, 6, 3, 10, 14, 4, 5, 3, 17, 9, 8, 9, 5, 3, 28, 21, 4])

# Set up the figure
fig, axes = plt.subplots(3, 2, figsize=(12, 15))

# Normal Distribution Probability Plot
stats.probplot(data, dist="norm", plot=axes[0, 0])
axes[0, 0].set_title('Normal Probability Plot')

# Exponential Distribution Probability Plot
stats.probplot(data, dist="expon", plot=axes[0, 1])
axes[0, 1].set_title('Exponential Probability Plot')

# Gamma Distribution Probability Plot
stats.probplot(data, dist="gamma", sparams=(2,), plot=axes[1, 0])  # shape parameter chosen as 2
axes[1, 0].set_title('Gamma Probability Plot')

# Lognormal Distribution Probability Plot
stats.probplot(data, dist="lognorm", sparams=(0.954, 0, np.exp(np.mean(np.log(data)))), plot=axes[1, 1])
axes[1, 1].set_title('Lognormal Probability Plot')

# Weibull Distribution Probability Plot
stats.probplot(data, dist="weibull_min", sparams=(1.5,), plot=axes[2, 0])  # shape parameter chosen as 1.5
axes[2, 0].set_title('Weibull Probability Plot')

# Adjust layout
plt.tight_layout()

# Save the figure to a file
plt.savefig('probability_plots.png')

plt.show()
```
