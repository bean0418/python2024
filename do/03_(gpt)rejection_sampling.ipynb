```python
import numpy as np
import matplotlib.pyplot as plt

# Define the target distribution (unnormalized)
def target_distribution(x):
    return np.exp(-x**2 / 2)

# Define the proposal distribution (uniform)
def proposal_distribution(size=1):
    return np.random.uniform(-5, 5, size)

# Rejection sampling
def rejection_sampling(n_samples):
    samples = []
    M = 1.0  # Scaling factor, chosen based on max value of target/proposal ratio
    while len(samples) < n_samples:
        # Sample from the proposal distribution
        x = proposal_distribution()
        
        # Uniform random number for acceptance criterion
        u = np.random.uniform(0, M)
        
        # Acceptance check
        if u <= target_distribution(x) / M:
            samples.append(x)
    
    return np.array(samples)

# Generate samples using rejection sampling
n_samples = 10000
samples = rejection_sampling(n_samples)

# Plot the histogram of the samples
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g')

# Plot the target distribution
x = np.linspace(-5, 5, 1000)
plt.plot(x, target_distribution(x), 'r-', lw=2, label='Target Distribution')
plt.title('Rejection Sampling')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.show()
```
