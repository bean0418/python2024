```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the parameters
population_size = 100000  # Size of the population
sample_size = 30          # Size of each sample
n_samples = 1000          # Number of samples to draw

# Generate a population with an unknown distribution (e.g., exponential distribution)
population = np.random.exponential(scale=4, size=population_size)

# Collect sample means
sample_means = []

for i in range(n_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    if i%100 == 0:
        print(f'{i}번째 sample mean은 {round(np.mean(sample), 3)}입니다.')
    sample_means.append(np.mean(sample))

# Plot the initial population distribution
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(population, kde=True, color='red')
plt.title('Population Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the distribution of the sample means
plt.subplot(1, 2, 2)
sns.histplot(sample_means, kde=True, color='blue')
plt.title(f'Distribution of Sample Means (Sample Size = {sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```
