```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the parameters
population_size = 100000  # Size of the population
n_samples = 1000          # Number of samples to draw

# Generate a population with an unknown distribution (e.g., exponential distribution)
population = np.random.exponential(scale=2, size=population_size)

# Sample sizes to compare
sample_sizes = [5, 20, 1000]

# Initialize the plot
plt.figure(figsize=(18, 10), dpi=1000)

# Plot the initial population distribution
plt.subplot(2, 2, 1)
sns.histplot(population, kde=True, color='red')
plt.title('Population Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the distributions of the sample means for different sample sizes
for i, sample_size in enumerate(sample_sizes, start=2):
    sample_means = []

    for j in range(1, n_samples+1):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))
        if j % 500 == 0:
            print(f'{i-1}, {j}th sample')

    plt.subplot(2, 2, i)
    sns.histplot(sample_means, kde=True, color='blue')
    plt.title(f'Distribution of Sample Means (Sample Size = {sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')

    if (i == 4) & (j == 1000):
        print("plotting....")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
